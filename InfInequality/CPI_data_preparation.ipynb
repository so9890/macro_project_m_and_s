```python
data.loc[:,'series_id'] =  pd.Series(data.T.loc['series_id']).str.cat(additional_column.T.loc['suffix'], sep='', na_rep=' ')
```

```python
boolean_isnan = pd.isna(pd.Series(additional_column.T.loc['year']))
```

```python
## the below code does not give an error but takes ages to run.
## what i had before worked faster but gave a warning: 
## data.loc[:,'year'][boolean_isnan == False] = additional_column.loc[:,'year']
```

```python
 data.loc[:,'year'][boolean_isnan == False] = additional_column.loc[:,'year']
```

```python
for i in range(len(boolean_isnan)):
    if boolean_isnan[i] == False:
          data.loc[i,'year'] = additional_column.loc[i,'year']
```
